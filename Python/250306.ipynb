```python
# 매개변수 list를 그대로 사용하면 저장하지 않아도 바뀐다
def replace_list(list, a, b):
    for i in range(len(list)):
        list[i] = b if list[i] == a else list[i]
```

```python
def replace_object(value, a, b):
    result = ""
    if type(value) == str:
        i = 0
        while i < len(value):        
            if value[i:i+len(a)] == a:
                result += b
                i += len(a)
            else:
                result += value[i]
                i+=1
    else:
        for i in range(len(value)):
            result.append(b if value[i] == a else value[i])
            
    return result
```

```python
df['item_name'] = df['item_name'].str.upper().str.replace(" ","")
all_item = df['item_name'].unique()
price_dict = df[['item_name', 'item_price']].groupby(['item_name']).mean()
price_null = df['item_price'].isna()
for item in all_item:
    name_a = df['item_name'] == item
    price_mean = price_dict.loc[item]
    df.loc[name_a & price_null,'item_price'] = price_mean
```
