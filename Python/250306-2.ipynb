```python
import pandas as pd
```

```python
df = pd.read_csv("./data/csv/Sales Records.csv")
df.head()
```

```python
df.drop(['Order ID','Total Revenue','Total Cost','Total Profit','Order Priority'], axis=1, inplace=True)
```

```python
column_names = ['대륙','국가','상품종류','판매채널','주문일자','발송일자','판매개수','단가','원가']
df.columns = column_names
```

```python
df.head()
```

```python
df.sort_values(['대륙','국가'],inplace=True)
```

```python
df.head()
```

```python
df['총단가'] = df['판매개수'] * df['단가']
df['총원가'] = df['판매개수'] * df['원가']
df['총이윤'] = df['총단가'] - df['총원가']
```

```python
df.head()
```

```python
df.reset_index(drop=True, inplace=True)
```

```python
df.head()
```

```python
df.groupby('국가').sum()
```

```python
df.groupby('국가')[['총이윤']].sum()
```

```python
group_data = df.groupby('국가')[['총이윤']].sum()
group_data.sort_values('총이윤', ascending=False, inplace=True)
```

```python
group_data['rank'] = [i+1 for i in range(len(group_data))]
```
