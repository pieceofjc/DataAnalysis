```python
trade_20 = boll.create_trade(
    boll.create_band(df_aapl, _col = 'Close', _start = '2022-01-01')
)

trade_30 = boll.create_trade(
    boll.create_band(df_aapl, _col = 'Close', _start = '2022-01-01', _cnt = 30)
)
```

```python
trade_20.head()
trade_30.head()
```

```python
trade_20 = trade_20[['Close', 'trade']]
trade_30.drop(['center', 'up', 'down'], axis=1, inplace= True)
```

```python
import pandas as pd
```

```python
trade_2030 = pd.concat([trade_20, trade_30], axis = 1, ignore_index=True)
```

```python
trade_2030.drop([2], axis=1, inplace=True)
```

```python
trade_2030.columns = ['Close', 'trade20', 'trade30']
```

```python
trade_2030[['Close']]
```

```python
flag = (trade_2030['trade20'] == 'buy') & (trade_2030['trade30'] == 'buy')
t1 = trade_2030.copy()
t1.loc[flag, 'trade'] = 'buy'
t1
```
