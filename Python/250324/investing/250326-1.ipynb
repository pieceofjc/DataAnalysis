```python
# case3
# unique() 함수를 이용

ym_list = df['STD-YM'].unique()
# 빈데이터프레임 생성
month_last_df = pd.DataFrame()
for ym in ym_list:
    data = df.loc[ym].tail(1)
    month_last_df = pd.concat([month_last_df, data], axis=0)
```

```python
# month_last_df에서
# 전월의 수정 종가 데이터를 BF_1M에 대입
# 전년도의 수정종가 데이터를 BF_12M에 대입

month_last_df['BF_1M'] = month_last_df['Adj Close'].shift(1)
month_last_df['BF_12M'] = month_last_df['Adj Close'].shift(12)
```

### 거래 내역 생성
- 전월의 수정종가가 전년도의 수정종가 보다 높을 때 매수한다
- 모멘텀 인덱스 : (전월의 수정종가 / 전년도의 수정종가) - 1
- 모멘텀 인덱스가 0보다는 크고 무한대가 아닌 경우 -> 매수

```python
# 거래내역은 df에 생성
df['trade'] = ''
for idx in month_last_df.index:
    signal = ''
    momentum_index = month_last_df.loc[idx, 'BF_1M'] / month_last_df.loc[idx, 'BF_12M'] - 1
    flag = (momentum_index > 0) & (momentum_index != np.inf)
    if flag:
        signal = 'buy'
    
    print(f"날짜 : {idx}, 모멘텀 인덱스 : {momentum_index}, signal : {signal}")
    df.loc[idx:, 'trade'] = signal
```

```python
# 모멘텀 인덱스가 1보다 작을 때는 전년도의 값보다 전월의 값이 작을 때
df['trade'].value_counts()
```

```python
# 수익률
df2 = df.copy()
df2['rtn'] = 1
for idx in df2.index:
    # 매수
    if (df2.shift().loc[idx, 'trade'] == '') & (df2.loc[idx,'trade'] == 'buy'):
        buy = df2.loc[idx, 'Adj Close']
        print(f"매수일 : {idx}, 매수가 : {buy}")
    elif (df2.shift().loc[idx, 'trade'] == 'buy') & (df2.loc[idx, 'trade'] == ''):
        sell = df2.loc[idx, 'Adj Close']
        print(f"매도일 : {idx}, 매도가 : {sell}")
        
        rtn = sell / buy
        df2.loc[idx, 'rtn'] = rtn
        print(f"수익률 : {rtn}")
    
df2['acc_rtn'] = df2['rtn'].cumprod()
acc_rtn = df2.iloc[-1,-1]
print(acc_rtn)
```

    - 구매내역이 추가된 데이터프레임을 되돌려준다


```python
# STD-YM 컬럼 만들기

def create_ym(
    _df,
    _col = 'Adj Close'
):
    result = _df.copy()
    if 'Date' in result.columns:
        result.set_index('Date', inplace=True)
        
    result.index = pd.to_datetime(result.index, utc = False)
    
    flag = result.isin([np.nan, np.inf, -np.inf]).any(axis=1)
    result = result.loc[~flag, [_col]]
    
    result = result[[_col]]
    
    result['STD-YM'] = result.index.strftime('%Y-%m')
    
    return result
```

```python
new_df = pd.read_csv("../../data/csv/AMZN.csv", index_col = "Date")
newnew = create_ym(new_df)
```

```python
def create_month(
    _df,
    _start = '2010-01-01',
    _end = datetime.now(),
    _momentum = 12,
    _select = 1
):
    
    if _select == 1:
        flag = _df['STD-YM'] != _df['STD-YM'].shift(-1)
    elif _select == 0:
        flag = _df['STD-YM'] != _df['STD-YM'].shift(1)
    else:
        return "_select의 값은 0 또는 1입니다"
    
    result = _df.loc[flag]
    
    col = result.columns[0]
    
    result['BF1'] = result[col].shift(1)
    
    result['BF2'] = result[col].shift(_momentum)
    
    try:
        result.index = result.index.tz_localize(None)
        start = datetime.strptime(_start, '%Y-%m-%d')
    except Exception as e:
        print(e)
    
    
    result = result.loc[start:_end]
    return result
        
    
        
```
