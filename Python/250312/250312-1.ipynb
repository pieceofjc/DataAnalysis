```python
section_sise_top = soup2.find(class_="section_sise_top")
tbl_home = section_sise_top.find(class_="tbl_home")
tr_list = tbl_home.find_all("tr")

result_list = []
for i in range(1, len(tr_list)):
    result = []
    result.append(tr_list[i].find("th").get_text().strip())
    result += list(map(lambda x : x.get_text().strip() ,tr_list[i].find_all("td")))
    result_list.append(result)
    
result_list
    
```

```python
def func_1(tr_list):
    result_list = []
    for i in range(0, len(tr_list)):
        result_list.append(
            list(
                map(
                    lambda x : x.get_text().strip(),
                    tr_list[i].find_all(["th", "td"])
                )
            )
        )
        
    return result_list
```

```python
data = func_1(tr_list)
```

```python
import pandas as pd
col = data[0]
data = data[1:]
df = pd.DataFrame(data, columns=col)
```
