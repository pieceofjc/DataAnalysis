### 데이터프레임의 필터링
- df.loc[인덱스의 조건, 컬럼의 조건]
    - 인덱스의 값을 기준으로 필터
    - 컬럼의 값을 기준으로 필터
- df.iloc[인덱스의 위치, 컬럼의 위치]
    - 인덱스의 위치를 기준으로 필터

- df[컬럼명]
    - 특정 컬럼의 데이터를 확인
    - 결과가 Series 출력
- df[[컬럼명]]
    - 특정 컬럼의 데이터를 확인
    - 결과가 Dataframe 출력력?

```python
# 일일확진자 일일사망자 컬럼을 추가하는 함수
def add_column(ddf, target):
    df = ddf.copy()
    # 특정 컬럼 제거
    df.drop(
        ["Unnamed: 0", "seq"],
        axis = 1,
        inplace=True
    )
    
    # 컬럼의 이름을 변경
    column_names = ['등록일시', '총사망자', '총확진자', '기준일', '기준시간', '수정일시', '누적의심자', '누적확진율']
    df.columns = column_names
    
    # 특정 컬럼의 데이터를 기준으로 오름차순 정렬
    df.sort_values(target, inplace=True)
    
    # 인덱스 초기화
    df.reset_index(drop=True, inplace=True)

    # 일일확진자 컬럼을 추가
    df['일일확진자'] = df['총확진자'].diff(1).fillna(0)
    # 일일사망자 컬럼을 추가
    df['일일사망자'] = df['총사망자'].diff(1).fillna(0)

    return df
```

```python
df = pd.read_csv("./data/csv/corona.csv")
new_df = add_column(df, '기준일')
```
